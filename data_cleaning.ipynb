```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('top5leagues.csv')
df.head()
```

```python

duplicates=df[df.duplicated(subset=['Player'], keep=False)]
print(duplicates['Player'].unique())
```

```python
duplicates = df[df['Player'].isin(duplicates['Player'].unique())]
duplicates.head(30)
```

```python
df=df[df['Player']!='Player']
```

```python
df.set_index('Rk', inplace=True)
df.head(50)
```

```python

df.drop(columns='Matches',axis=1,inplace=True)
df.columns
```

```python
df['Nation'].isnull().value_counts()
df[df['Nation'].isnull()]
```

```python
df.loc['1349','Nation']="ESP"
df.loc['1166','Nation']="TUR"
df.loc['1706','Nation']="FRA"
```

```python
import re
def nation_fix(nation):
    nat_fixed=re.findall(r'[A-Z]',nation)
    nat_fixed=''.join(nat_fixed)
    return nat_fixed

vec_func=np.vectorize(nation_fix)
df['Nation']=vec_func(df['Nation'])
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
def comp_fix(comp):
    result = re.sub(r'^[a-z\s]+', '', comp)
    return result
df['Comp']=df['Comp'].apply(comp_fix)
```

```python
def age_fix(age):
    age=str(age)
    result = re.sub(r'-.*', '', age)
    return result
df['Age']=df['Age'].apply(age_fix).astype(int)
```

```python
df.head()
```

```python
df.to_csv('top5leagues_final.csv')
```
