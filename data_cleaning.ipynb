```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('top5leagues-1.csv')
df.head()
```

```python

duplicates=df[df.duplicated(subset=['Player'], keep=False)]
print(duplicates['Player'].unique())
```

```python
duplicates = df[df['Player'].isin(duplicates['Player'].unique())]
duplicates.head(30)
```

```python
df=df[df['Player']!='Player']
```

```python
df.set_index('Rk', inplace=True)
df.head(50)
```

```python

df.drop(columns='Matches',axis=1,inplace=True)
df.columns
```

```python
import re
def nation_fix(nation):
    nat_fixed=re.findall(r'[A-Z]',nation)
    nat_fixed=''.join(nat_fixed)
    return nat_fixed

vec_func=np.vectorize(nation_fix)
df['Nation']=vec_func(df['Nation'])
```

```python
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
df.to_csv('top5leagues_final.csv')
```
